```python
#NumPy array
import numpy as np
#integer array
np.array([1,5,3,8,4])
```

```python
#Automatic/Implicit conversion of values/elements
np.array([10.1,2,8,4,5])
```

```python
#Explicit conversion of values/elements
np.array([1,2,5,4,3], dtype = 'float32')
```

```python
#[[,,],[,,],[,,]]
#nested list result in multi-dimensional array
np.array([range(i,i+3) for i in [1,4,7]])
```

```python
#Create a length in integer array filled with zeros
np.zeros(10,dtype=int)
```

```python
#Create array of zero of float type
np.zeros((2,5),dtype=float)
```

```python
#Create array of one of float type
np.ones((2,5),dtype=float)
```

```python
#Create a 3x5 array filled with 3.14
np.full((3,5),3.14)
```

```python
 
# Create an array filled with a linear sequence
# Starting at 0, ending at 20, stepping by 3
# (this is similar to the built-in range() function)
np.arange(0, 20, 3)
```

```python
#0 to 2   4 element   
#Creation of an array of evenly spaced values for the given range
#0 is starting, 1 is destination and 3 is steps/elements
np.linspace(0, 1, 3)
```

```python
# Create a 3x3 array of uniformly distributed
# random values between 0 and 1
np.random.random((3, 3))
```

```python
# Create a 3x3 array of normally distributed random values
# with mean 0 and standard deviation 1
np.random.normal(0, 1, (2, 3))  # 0 is mean 1 is SD and 2,3 is dimension(row & column following)
```

```python
# Create a 3x3 array of random integers in the interval [0, 10)
np.random.randint(0, 10, (3, 3)) # 0 is starting 10 is ending but not include and last one is 3,3 is dimension(rows and columns )
```

```python
# Create a 3x3 identity matrix
np.eye(3)
```

```python
# Create an uninitialized array of three integers
# The values will be whatever happens to already exist at that memory location
np.empty(3)
```

#**Numpy**

```python
import numpy as np
np.random.seed(20)  # seed for reproducibility
x1 = np.random.randint(20, size=10)  # One-dimensional array
x1
```

```python
#Seed 100 value can go upto 200 and create a array as 5,6,7(3x3 dimension)
np.random.seed(100)
x2 = np.random.randint(200, size=(5,6,7)) #->5 is SETs, ->6 is ROWs & ->7 is COLUMNs
x2
```

```python
#Seed 100 value can go upto 200 and create a array as 5,6,7(3x3 dimension)
np.random.seed(100)
x2 = np.random.randint(200, size=(5,2,2)) #->5 is SETs, ->2 is ROWs & ->2 is COLUMNs
x2
```

```python
np.random.seed(100)
x2 = np.random.randint(200, size=(4,2,2,2)) #->4 is repeted,->2 is SETs, means =>> 4 * 2 = 8 ->6 is ROWs & ->7 is COLUMNs (4x4 dimension)
x2
```

```python
# 5 dimensions 3x4x3x2x2
np.random.seed(100)
x2 = np.random.randint(200, size=(3,4,3,2,2)) #->3 times is repeted,->2 is SETs, means =>> 4 * 2 = 8 ->6 is ROWs & ->7 is COLUMNs (4x4 dimension)
x2
```
